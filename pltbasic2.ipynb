```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.3)
#plt.fill_between(x=[2,5],y1=2,y2=4)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,3,4,1]

plt.title("Python",fontsize=15)
plt.xlabel("Growth",fontsize=10)
plt.ylabel("Days",fontsize=10)
plt.text(1,3,"graph",fontsize=10,style="italic",bbox={"facecolor":"pink"})
plt.annotate("there",xy=(2,1),xytext=(3,3),arrowprops={"facecolor":"black","shrink":1})

plt.plot(x,y,color="m")
plt.legend(["increase"],loc=9,facecolor="cyan",edgecolor="r",shadow=True,framealpha=0.3)
plt.show()
```
