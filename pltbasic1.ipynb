```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,2,5,6,4]
plt.stem(x,y,linefmt=":",basefmt="g",label="python",bottom=0,markerfmt="r*",orientation="vertical")
plt.legend()
plt.show()
#use_line_collection=False  not working in this version
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,notch=False,vert=True,widths=0.2,labels=["python"],patch_artist=True,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="m"),whiskerprops=dict(color="y"))
plt.show()
#in boxplot always give label in list
#if we gives no more than 70 then it will be outlier and will be represnted by circle....try with 120
#whis = 2     joins with the outlier
#sym = "g+"   for outlier
#flierprops=dict(markeredgecolor="g")   for outlier
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,notch=False,vert=True,widths=0.2,labels=["python"],patch_artist=True,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="m"),whiskerprops=dict(color="y"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
y = [10,20,30,50,70,80,90]
z = [x,y]
plt.boxplot(z,labels=["python","c++"])
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,5,4,5]
area2 = [3,4,5,2,1]
area3 = [2,5,4,2,3]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=[
    "m","r","y"],baseline="zero")
plt.legend(loc=2)
plt.show()
#baseline="wiggle"/"zero"/"sym"
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.step(x,y,marker="o",ms=10,mfc="r")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```
