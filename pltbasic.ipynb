```python
import matplotlib.pyplot as plt
x = ["python","c++","java","c"]
y = [70,90,80,75]
plt.xlabel("languages",fontsize=10)
plt.ylabel("No.")
plt.title("Information")
c = ["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)  #align="edge"/"center" alpha value should be bet 0 to 1
plt.show()
```

```python
x = ["python","c++","java","c"]
y = [70,90,80,75]
plt.xlabel("languages",fontsize=10)
plt.ylabel("No.")
plt.title("Information")
c = ["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",label="computer")
plt.legend()  #for label parameter  
plt.show()
```

```python
x = ["python","c++","java","c"]
y = [70,90,80,75]
z = [20,30,15,25]
plt.xlabel("languages",fontsize=10)
plt.ylabel("No.")
plt.title("Information")
plt.bar(x,y,width=0.4,color="r",label="computer")
plt.bar(x,z,width=0.4,color="y",label="popularity")
plt.legend()  #for label parameter  
plt.show()
```

```python
import numpy as np
x = ["python","c++","java","c"]
y = [70,90,80,75]
z = [20,30,15,25]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("languages",fontsize=10)
plt.ylabel("No.")
plt.title("Information")
plt.bar(p,y,width,color="r",label="computer")
plt.bar(p1,z,width,color="y",label="popularity")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()  #for label parameter  
plt.show()
```

```python
x = ["python","c++","java","c"]
y = [70,90,80,75]
z = [20,30,15,25]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.ylabel("languages",fontsize=10)
plt.xlabel("No.")
plt.title("Information")
plt.barh(p,y,width,color="r",label="computer")
plt.barh(p1,z,width,color="y",label="popularity")

plt.yticks(p+width/2,x,rotation=10)
plt.legend()  #for label parameter  
plt.show()
```

```python
x = [10,20,30,40]
y = ["c","java","c++","python"]
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow="True",labeldistance=1.1,radius=0.8,textprops={"fontsize":10},wedgeprops={"linewidth":3})
plt.show()
#startangle=180  rotates whole pie chart
#rotatelabels=True
```
